```python
import lcapy
from lcapy import Circuit

from IPython.display import display, Latex
%matplotlib inline
```

```python
sch='''
VS 1 0 ; down
W 1 2 ;  right, size=2
R1 2 3 ; down
R2 3 4; down
W 3 5; right
P1 5 6; down,v=V_{out}
W 4 6; right
W 4 0; left
'''

#Demonstrate thate we can write the descriptioon to a file 
fn="voltageDivider.sch"
with open(fn, "w") as text_file:
    text_file.write(sch)

#  and then create the circuit model from the (persisted) file    
cct = Circuit(fn)
```

```python
cct.draw(style='american', draw_nodes=False, label_nodes=False)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import sklearn
```

```python
removed = cv2.imread('removed.jpeg')
removed = cv2.cvtColor(removed, cv2.COLOR_RGB2GRAY)
img = cv2.imread('test_images/test8.jpeg')
plt.imshow(removed)
```

```python
fast = cv2.FastFeatureDetector_create(threshold=1, nonmaxSuppression=False, type=cv2.FAST_FEATURE_DETECTOR_TYPE_9_16)
kp = fast.detect(removed, None)
```

```python
img2 = img.copy()
img2 = cv2.drawKeypoints(img2, kp, None, color=(255,0,0))
plt.imshow(img2)
```

```python
anchor_points = cv2.goodFeaturesToTrack(image=removed,maxCorners=100,qualityLevel=0.4,minDistance=60)
```

```python
img3 = img.copy()
for i in anchor_points:
    cv2.circle(img3, (int(i[0][0]), int(i[0][1])), 5, color=(255,0,0), thickness=10)
    
plt.imshow(img3)
```
